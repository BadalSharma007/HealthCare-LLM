```python
import torch  # core DL library
from transformers import AutoTokenizer, AutoModelForCausalLM  # HF tools

model_id = "google/medgemma-1.5-4b-it"  # model name on Hugging Face
token = os.environ.get("HF_TOKEN", "")  # your HF access token

# Load tokenizer (text → tokens)
tokenizer = AutoTokenizer.from_pretrained(
    model_id,
    token=token  # needed for gated model
)

# Load model (weights)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16  # use FP16 → lower memory
)

# Move model to Apple GPU (M3)
model = model.to("mps")  # MPS = Apple Metal GPU

# Input prompt
prompt = "I have fever and headache, what should I do?"

# Convert text → tensors
inputs = tokenizer(prompt, return_tensors="pt")

# Move input to same device as model
inputs = inputs.to("mps")

# Generate output
outputs = model.generate(
    **inputs,
    max_new_tokens=20  # limit response length
)

# Convert tokens → readable text
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
